```python
def scrap_all_categories(max_companies=2, max_reviews=2, max_req=-1):
    df_categories = pd.read_csv("csv/categories.csv")
    df_categories.head()
    df_big_categories_link = df_categories[df_categories["category"] == 0]["link"]
    companies = {"name": [], "link": [], "stars": [], "review_count": [], "category": [], "location": []}
    reviews = {"consumer_name": [], "consumer_nb_review_writed": [], "stars": [],
               "title_review": [], "content_review": [], "date_experience": [], "company": [], "category": []}


    #40, 240
    for i in tqdm(range(len(df_big_categories_link))):
    #for i in range(len(df_big_categories_link)):
        new_companies, new_reviews = scrap_companies_and_reviews_from_category(df_big_categories_link[i], max_companies=max_companies,
                                                                               max_reviews=max_reviews, verbose=0, max_req=max_req)

        for key in companies:
            companies[key] += new_companies[key]
        for key in reviews:
            reviews[key] += new_reviews[key]

    df_companies = pd.DataFrame(data=companies)
    df_reviews = pd.DataFrame(data=reviews)
    return df_companies, df_reviews
```

```python
def scrap_all_companies(max_req=-1):
    df_categories = pd.read_csv("csv/categories.csv")
    df_categories.head()
    df_big_categories_link = df_categories[df_categories["category"] == 0]["link"]
    companies = {"name": [], "link": [], "stars": [], "review_count": [], "category": []}
    for i in tqdm(range(len(df_big_categories_link))):
        new_companies = scrap_companies_from_category(df_big_categories_link[i], max_companies=-1, max_req=max_req)
        for key in companies:
            companies[key] += new_companies[key]
    df_companies = pd.DataFrame(data=companies)
    return df_companies    
```

```python
df_companies, df_reviews = scrap_all_categories(2, 2)

print(df_companies.shape)
print(df_companies.info())
display(df_companies.head())

print(df_reviews.shape)
print(df_reviews.info())
display(df_reviews.head())

df_companies.to_csv("scraped_companies.csv", sep=";", index=False)
df_reviews.to_csv("scraped_reviews.csv", sep=";", index=False)
```
